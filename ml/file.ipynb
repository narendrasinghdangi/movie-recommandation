```python
import pandas as pd
import numpy as np
import turicreate as tc
```

```python
df = pd.read_excel("data/SpareItemDump.xlsx",engine="openpyxl")
```

```python
df.dropna(subset = ["PurchaseOrderId"], inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.dropna(subset = ["Make"], inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df["PurchaseOrderId"]=df["PurchaseOrderId"].astype(int)
```

```python
d={}
for i in range(len(df)):
    if df["PurchaseOrderId"][i] in d:
        if df["product"][i] in d[df["PurchaseOrderId"][i]]:
            (d[df["PurchaseOrderId"][i]])[df["product"][i]] +=1
        else:
            (d[df["PurchaseOrderId"][i]])[df["product"][i]] =1
            
    else:
        d[df["PurchaseOrderId"][i]]={df["product"][i]:1}
```

```python
d
```

```python
li=[]
for i in d:
    for j in d[i]:
        li.append([i,j,d[i][j]])
data = pd.DataFrame(li,columns=["PurchaseOrderId","product","purchase_count"])
```

```python
data
```

```python
def create_data_dummy(data):
    data_dummy = data.copy()
    data_dummy['purchase_dummy'] = 1
    return data_dummy
data_dummy = create_data_dummy(data)
```

```python
data_dummy.to_csv("my_data.xlsx")
```

```python
df_matrix = pd.pivot_table(data, values='purchase_count', index='PurchaseOrderId', columns='product')
df_matrix_norm = (df_matrix-df_matrix.min())/(df_matrix.max()-df_matrix.min())
# create a table for input to the modeling  
d = df_matrix_norm.reset_index()
d.index.names = ['scaled_purchase_freq'] 
data_norm = pd.melt(d, id_vars=['PurchaseOrderId'], value_name='scaled_purchase_freq').dropna()
print(data_norm.shape)
data_norm
```

```python
unique_garage = []

for i in data["Garage"]:
    if i not in unique_garage:
        unique_garage.append(i)
```

```python
data = tc.SFrame(data)
data_dummy =  tc.SFrame(data_dummy)
data_norm = tc.SFrame(data_norm)
```

```python
user_id = 'PurchaseOrderId'
item_id = 'product'
n_rec = 3
n_display = 30 # to display the first few rows in an output dataset
```

```python
final_model = tc.item_similarity_recommender.create(tc.SFrame(data_dummy),
                                            user_id=user_id, 
                                            item_id=item_id,
                                            target='purchase_dummy', similarity_type='cosine',verbose=0)
```

```python
def create_output(model, user_to_recommend, n_rec):
    recomendation = model.recommend(users=user_to_recommend, k=n_rec)
    return recomendation
```

```python
user_to_recommend_item =[123]
df_output = create_output(final_model,user_to_recommend_item , n_rec)
```
