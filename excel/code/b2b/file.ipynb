```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("K_weeklyreport_Oct01_Oct30.csv")
```

```python
df
```

```python
df=df.assign(new_data=None)
```

```python
for i in range(len(df)):
    df["new_data"][i] = str(df["request_id"][i])+"_"+str(df["request_item_id"][i])
```

```python
ind=[]
uni=[]
for i in range(len(df)):
    if df["new_data"][i] not in uni:
        uni.append(df["new_data"][i])
        ind.append(i)
```

```python
universal_data=df.iloc[ind]
```

```python
universal_data=universal_data.reset_index()
universal_data.drop(["index"],axis=1,inplace=True)
```
