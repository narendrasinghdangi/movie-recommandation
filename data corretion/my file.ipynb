```python
varient_key_need = []
for i in range(len(df)):
    if str(df["first_set"][i]).lower() in varinet_key_data:
        df["variant_key"][i]=varinet_key_data[str(df["first_set"][i]).lower()]
    else:
        varient_key_need.append(df["first_set"][i])
```

```python
unique_list = []
for x in varient_key_need:
        if x not in unique_list:
            unique_list.append(x)
```

```python
len(unique_list)
```

```python
varinet_need = pd.DataFrame()
```

```python
varinet_need["boodmo"] =unique_list
```

```python
mat1 = []
mat2 = []
p = []

list1 = varinet_need["boodmo"].tolist()
list2 = df_varient_key['Variant'].tolist()
  
threshold = 80

for i in tqdm(range(len(list1))):
    mat1.append(process.extract(str(list1[i]), list2, limit=1))
varinet_need['matches'] = mat1

```

```python
for j in varinet_need['matches']: 
    for k in j: 
        if k[1] > threshold: 
            p.append(k[0])
              
    mat2.append(",".join(p)) 
    p = [] 
varinet_need['matches'] = mat2
```

```python
varinet_need=varinet_need.assign(varient_key=None)
```

```python
varient_dic={}
for i in range(len(df_varient_key)):
    varient_dic[df_varient_key["Variant"][i]]=df_varient_key["variant_key"][i]
```

```python
for i in range(len(varinet_need)):
    if varinet_need["matches"][i] in varient_dic:
        varinet_need["varient_key"][i]=varient_dic[varinet_need["matches"][i]]
```
