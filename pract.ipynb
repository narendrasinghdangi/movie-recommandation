```python
import numpy as np
import pandas as pd
```

```python
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits
```

```python
movies
```

```python
credits_rename=credits.rename(columns={"movie_id":"id"})
```

```python
credits_rename
```

```python
movies_merge=movies.merge(credits_rename,on="id")
```

```python
movies_merge.head()
```

```python
movies_merge_dropped=movies_merge.drop(columns=["homepage","title_x","title_y","status","production_companies"])
```

```python
movies_merge_dropped
```

```python
movies_merge_dropped.info()
```

```python
movies_merge_droppeda=movies_merge_dropped.drop(columns=["spoken_languages","original_language","runtime","release_date","production_countries"])
```

```python
movies_merge_droppeda.head()
```

```python
v=movies_merge_droppeda["vote_count"]
```

```python
R=movies_merge_droppeda["vote_average"]
```

```python
C=movies_merge_droppeda["vote_average"].mean()
```

```python
C
```

```python
m=movies_merge_droppeda["vote_count"].quantile(0.7)
```

```python
m
```

```python
movies_merge_droppeda["weighted_average"]=((R*v)+(C*m))/(v+m)
```

```python
movies_merge_droppeda
```

```python
movies_sorted_ranking=movies_merge_droppeda.sort_values("weighted_average",ascending=False)
```

```python
movies_sorted_ranking[["original_title","popularity","vote_average","weighted_average"]]
```
