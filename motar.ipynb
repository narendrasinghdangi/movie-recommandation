```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv("motrparts.csv")
```

```python
df
```

```python
df1=pd.DataFrame()
```

```python
for i in range(len(df)):
    li=[]
    part_no = str(df["Part Number"][i]).split(":")[1][1:]
    sp=part_no.split("-")
    if len(sp)!=2:
        part_no=str(part_no[:5])+"-"+str(part_no[5:])
        
    try:
        comp = str(df["compatibility"][i]).split("|")
        year = comp[3]
        model = comp[0]
        
        engine = comp[1]
        fuel = comp[2]
    except:
        model =""
        year = ""
        engine = ""
        fuel = ""
    
    li.append([part_no,df["Parts Brand"][i],"",df["desc1"][i],float(df["MRP"][i]),df["Parts Category"][i],
               "","",df["Car Make"][i],str(model),str(year),str(engine),"",str(fuel),"",df["weight"][i],
               df["Dimensions"][i],df["Net Quantity"][i]])
    df2 = pd.DataFrame(li)
    df1 = pd.concat([df1,df2])
```
